```python
from tkinter import *
from fpdf import FPDF

# Função para gerar o PDF
def gerar_orcamento():
    # Verifica se todos os campos foram preenchidos
    if not projeto_entry.get() or not horas_entry.get() or not valor_entry.get() or not prazo_entry.get():
        mensagem_label.config(text="Por favor, preencha todos os campos")
        return
    
    projeto = projeto_entry.get()
    horas_estimada = horas_entry.get()
    valor_hora = valor_entry.get()
    prazo = prazo_entry.get()

    try:
        # Verifica se os valores digitados são números válidos
        horas_estimada = int(horas_estimada)
        valor_hora = float(valor_hora)
        prazo = int(prazo)
    except ValueError:
        mensagem_label.config(text="Por favor, digite números válidos")
        return

    valor_total_estimado = horas_estimada * valor_hora

    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Helvetica")

    pdf.image("template.png", x=0, y=0)

    pdf.text(115, 145, projeto)
    pdf.text(115, 160, str(horas_estimada))
    pdf.text(115, 175, f"R${valor_hora:.2f}")
    pdf.text(115, 190, f"{prazo} dias")
    pdf.text(115, 205, f"R${valor_total_estimado:.2f}")

    pdf.output("pdf_projeto.pdf")
    mensagem_label.config(text="Orçamento gerado com sucesso")

# Cria a janela
janela = Tk()
janela.title("Gerador de orçamentos")
janela.geometry("400x300")

# Cria os campos de entrada
projeto_label = Label(janela, text="Descrição do projeto:")
projeto_label.pack()
projeto_entry = Entry(janela)
projeto_entry.pack()

horas_label = Label(janela, text="Horas estimadas:")
horas_label.pack()
horas_entry = Entry(janela)
horas_entry.pack()

valor_label = Label(janela, text="Valor da hora (em R$):")
valor_label.pack()
valor_entry = Entry(janela)
valor_entry.pack()

prazo_label = Label(janela, text="Prazo (em dias):")
prazo_label.pack()
prazo_entry = Entry(janela)
prazo_entry.pack()

# Cria o botão para gerar o PDF
gerar_button = Button(janela, text="Gerar orçamento", command=gerar_orcamento)
gerar_button.pack()

# Cria a label para exibir a mensagem de sucesso
mensagem_label = Label(janela, text="")
mensagem_label.pack()

# Mantém a janela aberta
janela.mainloop()
```
